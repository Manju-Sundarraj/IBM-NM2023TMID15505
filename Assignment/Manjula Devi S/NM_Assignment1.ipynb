## 1. Download the dataset

The House Price Dataset of India is a dataset available on Kaggle that contains information on real estate properties in India. The dataset contains 23 columns,14620 rows and attributes such as,

1.Id(int)

2.Date(int)

3.number of bedrooms(int)

4.number of bathrooms(float)

5.living area(int)

6.lot area(int)

7.water front present(float)

8.number of views(int) etc..

## 2. Load the dataset

```python
import pandas as pd
data = pd.read_csv('/content/House Price India.csv')
print(data)
```

# 3.Perform the Below Visualizations.
### ● Univariate Analysis
### ● Bi - Variate Analysis
### ● Multi-Variate Analysis

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#task assigned in dataset
data['number of bathrooms'] = data['number of bathrooms'].astype(int)
data['waterfront present'] = data['waterfront present'].astype(int)
data = data.drop("Date" , axis=1)
```

###I. Univariate

```python
# Histogram for 'number of bedrooms'
sns.histplot(data['number of bedrooms'], bins=50,color='cyan')
plt.title('Number of Bedrooms Distribution')
plt.xlabel('Number of Bedrooms')
plt.ylabel('Frequency')
plt.show()
```

```python
# Boxplot for 'condition of house'
sns.boxplot(data['condition of the house'],color='m')
plt.title('Condition of House Distribution')
plt.xlabel('Condition of House')
plt.ylabel('Count')
plt.show()
```

```python
# Distribution plot for 'number of bathrooms'
sns.distplot(data['number of bathrooms'],bins=5,color='pink')
plt.title('Number of Bathrooms Distribution')
plt.xlabel('Number of Bathrooms')
plt.ylabel('Frequency')
plt.show()
```

```python
# Line plot for 'living area'
sns.lineplot(data['living area'])
plt.title('Living Area Distribution')
plt.xlabel('Living Area')
plt.ylabel('Frequency')
plt.show()
```

###II. Bi-Variate

```python
# Scatterplot between 'living area' and 'lot area'
sns.scatterplot(data=data, x='living area', y='lot area',color="y")
plt.title('Living Area vs. Lot Area')
plt.xlabel('Living Area')
plt.ylabel('Lot Area')
plt.show()
```

```python
# Scatterplot between 'number of bedrooms' and 'number of bathrooms'
sns.scatterplot(data=data, x='number of bedrooms', y='number of bathrooms',color='g')
plt.title('Number of Bedrooms vs. Number of Bathrooms')
plt.xlabel('Number of Bedrooms')
plt.ylabel('Number of Bathrooms')
plt.show()
```

###III. Multi-Variate

```python
#Pairplot for multiple attributes with 'water front present' as hue
sns.pairplot(data, vars=['number of bedrooms', 'living area', 'lot area', 'number of bathrooms'],
             hue='waterfront present', diag_kind='kde')
plt.suptitle('Pairplot of Attributes')
plt.show()
```

## 5. Handle the Missing values

####Ways to handle the Missing values
#####1.Dropping missing values.
#####2.Filling missing values with specified values.
#####3.Filling missing values using mean,median,mode.

```python
#Here we handle the missing data using mean,median and mode
data['number of bathrooms'] = data['number of bathrooms'].fillna(data['number of bathrooms'].mean())
data['waterfront present'] = data['waterfront present'].fillna(data['waterfront present'].median())
```
